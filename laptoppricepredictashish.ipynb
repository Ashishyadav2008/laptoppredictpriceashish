```python
# Memory column ko string me convert karo
laptops['Memory'] = laptops['Memory'].astype(str)

# '.0' ko hatao
laptops['Memory'] = laptops['Memory'].str.replace('\.0', '', regex=True)

# 'GB' aur 'TB' replace karo (TB ko 1000 banate hain)
laptops['Memory'] = laptops['Memory'].str.replace('GB', '', regex=False)
laptops['Memory'] = laptops['Memory'].str.replace('TB', '000', regex=False)

# '+' ke basis par split karo
new = laptops['Memory'].str.split('+', n=1, expand=True)

# Split ke baad 'first' aur 'second' columns banao
laptops['first'] = new[0].str.strip()
laptops['second'] = new[1].str.strip() if new.shape[1] > 1 else None

# 'first' me storage type check karo
laptops['Layer1HDD'] = laptops['first'].apply(lambda x: 1 if 'HDD' in x else 0)
laptops['Layer1SSD'] = laptops['first'].apply(lambda x: 1 if 'SSD' in x else 0)
laptops['Layer1Hybrid'] = laptops['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
laptops['Layer1Flash_Storage'] = laptops['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

# 'first' me sirf numbers rakho
laptops['first'] = laptops['first'].str.replace(r'\D', '', regex=True)

# Agar 'second' column empty hai, toh use '0' banao
laptops['second'] = laptops['second'].fillna('0')

# 'second' me storage type check karo
laptops['Layer2HDD'] = laptops['second'].apply(lambda x: 1 if 'HDD' in x else 0)
laptops['Layer2SSD'] = laptops['second'].apply(lambda x: 1 if 'SSD' in x else 0)
laptops['Layer2Hybrid'] = laptops['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
laptops['Layer2Flash_Storage'] = laptops['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

# 'second' me sirf numbers rakho
laptops['second'] = laptops['second'].str.replace(r'\D', '', regex=True)

# 'first' aur 'second' ko integer me convert karo
laptops['first'] = laptops['first'].astype(int)
laptops['second'] = laptops['second'].astype(int)

# Final storage columns calculate karo
laptops['HDD'] = (laptops['first'] * laptops['Layer1HDD'] + laptops['second'] * laptops['Layer2HDD'])
laptops['SSD'] = (laptops['first'] * laptops['Layer1SSD'] + laptops['second'] * laptops['Layer2SSD'])
laptops['Hybrid'] = (laptops['first'] * laptops['Layer1Hybrid'] + laptops['second'] * laptops['Layer2Hybrid'])
laptops['Flash_Storage'] = (laptops['first'] * laptops['Layer1Flash_Storage'] + laptops['second'] * laptops['Layer2Flash_Storage'])

# Intermediate columns ko drop karo
laptops.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                      'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                      'Layer2Flash_Storage'], inplace=True)
```

# Voting Regressor

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import VotingRegressor, GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Feature transformation step
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Initialize individual models
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)  # Added bootstrap=True
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=10, bootstrap=True)  # Added bootstrap=True

# Step 2: Voting Regressor combining all models
step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)], weights=[5, 1, 1, 1])

# Create pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline
pipe.fit(X_train, y_train)

# Make predictions
y_pred = pipe.predict(X_test)

# Evaluate model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

# Stacking

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import StackingRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Feature transformation step
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define the estimators for stacking
estimators = [
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5))
]

# Create the stacking regressor
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Make predictions on the test data
y_pred = pipe.predict(X_test)

# Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
